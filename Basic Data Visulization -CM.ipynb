## Load the Datasets

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

### Histogram

```python
plt.hist(train['SalePrice'])
plt.title("Sales Price Distribuion")
plt.xlabel("Value")
plt.ylabel("Frequency")
fig = plt.gcf()
```

```python
plt.hist(train['LotArea'])
plt.title("Lot Area Distribuion")
plt.xlabel("Value")
plt.ylabel("Frequency")
fig = plt.gcf()
```

```python
train['SalePrice'].plot.hist(grid=True, bins=20, rwidth=0.9,
                   color='#607c8e')
plt.title('Sales Price Distribuion')
plt.xlabel('Prices')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.75)
```

```python
sns.set_style('darkgrid')
sns.distplot(train['SalePrice'])
```

```python
num_data = train[['MSSubClass','LotFrontage','LotArea','PoolArea','MiscVal','MoSold','YrSold','SalePrice','SaleCondition']]
num_data
```

```python
num_data.head()
```

#### Visualizing the multidimensional relationships among the samples is as easy as calling sns.pairplot:

```python
sns.pairplot(num_data, hue='SaleCondition', size=2.2);
```

#### Faceted histograms

```python
grid = sns.FacetGrid(train, row="SaleCondition", col="YrSold", margin_titles=True)
grid.map(plt.hist, "SalePrice");
```

#### Factor plots

```python
with sns.axes_style(style='ticks'):
    g = sns.factorplot("YrSold", "SalePrice", "SaleCondition", data=train, kind="box")
    g.set_axis_labels("Sales Condition", "Sales Price");
```

#### Joint distributions

```python
with sns.axes_style('white'):
    sns.jointplot("SalePrice", "MoSold", data=train, kind='hex')
```

```python
with sns.axes_style('white'):
    sns.jointplot("SalePrice", "YrSold", data=train, kind='hex')
```

```python
sns.jointplot("SalePrice", "MoSold", data=train, kind='reg');
```

### Bar plots

```python
with sns.axes_style('white'):
    g = sns.factorplot("YrSold", data=train, aspect=2,
                       kind="count", color='steelblue')
    g.set_xticklabels(step=1)

```

```python
with sns.axes_style('white'):
    g = sns.factorplot("SaleCondition", data=train, aspect=2,
                       kind="count", color='steelblue')
    g.set_xticklabels(step=1)
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("SaleType", data=train, aspect=2,
                       kind="count", color='steelblue')
    g.set_xticklabels(step=1)
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("YrSold", data=train, aspect=4.0, kind='count',
                       hue='SaleCondition', order=range(2006, 2010))
    g.set_ylabels('Sales Condition Count')
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("MoSold", data=train, aspect=4.0, kind='count',
                       hue='SaleCondition')
    g.set_ylabels('Sales Condition Count')
```
