## FILTER, SORT, & GROUPBY

```python
# - Rows where the col column is greater than 0.5
mydf[mydf['agew'] > 50].sample(10)
```

```python
#- Rows where 40 > col > 50
mydf[(mydf['agew'] > 40) & (mydf['agew'] < 50)].sample(10) 
```

```python
mydf[(mydf['agew'] > 55) & (mydf['ageh'] > 50)].sample(10) 
```

```python
#- Sorts values by col1 in ascending order
mydf.sort_values('agew').head(10)
```

```python
# - Sorts values by col2 in descending order
mydf.sort_values('ageh',ascending=False).head()
```

```python
# - Sorts values by col1 in ascending order then col2 in descending order
mydf.sort_values(['agew','ageh'],ascending=[True,False]).head(10)
```

```python
# - Returns a groupby object for values from one column
mydf.groupby('work').count()
```

```python
# - Returns a groupby object values from multiple columns
mydf.groupby(['work','city']).count()
```

```python

#- Returns the mean of the values in col2, grouped by the
#values in col1 (mean can be replaced with almost any function from the statistics section)
mydf.groupby('work')['income'].mean().round(2)
```

```python
mydf.groupby('work')['experience'].mean().round(2)
```

```python
#- Creates a pivot table that groups by col1 and calculates the mean of col2 and col3
mydf.pivot_table(index='work',values=['income','experience'],aggfunc=np.mean).round(2) 
```

```python
mydf.pivot_table(index='work',values=['income','experience','educw','educh'],aggfunc=np.mean).round(2)
```

```python
#Transposed
mydf.pivot_table(index='work',values=['income','experience','educw','agew','educh','ageh'],aggfunc=np.mean).round(2).T
```

```python
# - Finds the average across all columns for every unique column 1 group
mydf.groupby('work').agg(np.mean).round(2)
```

```python
mydf.groupby('work').agg(np.mean).round(2).T
```

```python
mydf.groupby('city').agg(np.mean).round(2).T
```

```python
# - Applies a function across each column
mydf[mydf.columns[3:10]].apply(np.mean).round(2)
```

```python
#select all columns less work and city
mydf[mydf.columns.difference(['work', 'city'])].apply(np.mean).round(2)
```

```python
# - Applies a function across each row
mydf.apply(np.max, axis=0)
```
